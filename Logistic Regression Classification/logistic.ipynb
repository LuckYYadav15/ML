```python
data = data.drop(["PassengerId", "Name", "SibSp", "Parch", "Ticket", "Fare", "Cabin"], axis=1)
data.head()
```

```python
X_train = data[["Pclass", "Sex", "Age", "Embarked"]]
Y_train = data["Survived"]
```

```python
Y_train.head()
```

```python
X_train = X_train.to_numpy()
Y_train = Y_train.to_numpy()
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
print(X_train.shape[1])
print(X_train.shape[0])
```

```python
X = np.c_[np.ones(X_train.shape[0]), X_train]
X
```

```python
def model(X, Y, learning_rate, iterations):
    
    m = X_train.shape[0]
    n = X_train.shape[1]
    
    W = np.zeros((n+1,1))
    B = 0
    X = np.c_[np.ones(X.shape[0]), X]
    cost_list = []
    
    for i in range(iterations):
        
        Z = np.dot(W.T, X.T) 
        A = sigmoid(Z)
        
        # cost function
        cost = -(1/m)*np.sum( Y*np.log(A) + (1-Y)*np.log(1-A))
        
        # Gradient Descent
        dW = (1/m)*np.dot(A-Y, X)
        dB = (1/m)*np.sum(A - Y)
        
        W = W - learning_rate*dW.T
        # B = B - learning_rate*dB
        
        # Keeping track of our cost function value
        # print("W:", W)
        # print("B", B)
        cost_list.append(cost)
        
        if(i%(iterations/10) == 0):
            print("cost after ", i, "iteration is : ", cost)
        
    return W, cost_list
        
```

```python
iterations = 100000
learning_rate = 0.005
W, cost_list = model(X_train, Y_train, learning_rate = learning_rate, iterations = iterations)
```

```python
def pred(X, W):
  X = np.c_[np.ones(X.shape[0]), X]
  Z = np.dot(W.T, X.T) 
  A = sigmoid(Z)
  pred_list = []
  for x in list(A[0]):
    if x>=0.5:
      pred_list.append(1)
    else:
      pred_list.append(0)
  return pred_list
```

```python
pred_list = pred(X_train, W)
print(np.sum(np.square(np.array(pred_list) - Y_train)))
```
